# single dict

```python
from collections import defaultdict
def freqQuery(queries):
    dic = defaultdict(lambda:0)
    result = []
    for query in queries:
        if query[0]==1:
            dic[query[1]]+=1
        if query[0]==2:
            if dic[query[1]]>1:
                dic[query[1]]-=1 
        if query[0]==3:
            if query[1] in list(dic.values()):
                result.append(1)
            else:
                result.append(0)
    return result
```

```python
num = int(input())
queries = [0]*num
for i in range(num):
    queries[i] = [int(j) for j in input().strip().split()]
print(queries)
print(freqQuery(queries))
```
