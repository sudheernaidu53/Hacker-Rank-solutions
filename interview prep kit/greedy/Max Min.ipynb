# idea
1)sort the given array

2)initialise the minimum difference(ans) as a big number(infinity)

3)keep updating it in a for loop

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the maxMin function below.
def maxMin(k, arr):
    arr.sort()
    ans = float('inf')
    for i in range(len(arr)-k+1):
        print(i,'and', ans)
        ans = min(arr[i+k-1]-arr[i],ans)
    return ans
if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     n = int(input())

    k = int(input())

    arr = list(map(int,input().strip().split()))

#     for _ in range(n):
#         arr_item = int(input())
#         arr.append(arr_item)

    result = maxMin(k, arr)

#     fptr.write(str(result) + '\n')

#     fptr.close()
    print(result)
```
