```python
#!/bin/python3

import math
import os
import random
import re
import sys
# Complete the substrCount function below.
def substrCount(n, s):
    ans = 0
    freq = [0]*n
    i=0
    while i<n:
        j=i+1
        count=1
        while (j<n) and s[i]==s[j]:
            j+=1
            count+=1
        ans+=count*(count+1)/2
        freq[i]=count
        i = j
    for i in range(1,n):
        if s[i]==s[i-1]:
            freq[i]=freq[i-1]
    ## for odd length
    for i in range(1,n-1):
        if (s[i-1]==s[i+1]) and (s[i-1]!=s[i]):
            ans+=min(freq[i-1],freq[i+1])
    return int(ans)     
        

if __name__ == '__main__':
    s = input()
    n = len(s)
    result = substrCount(n, s)
    print(result)
```
