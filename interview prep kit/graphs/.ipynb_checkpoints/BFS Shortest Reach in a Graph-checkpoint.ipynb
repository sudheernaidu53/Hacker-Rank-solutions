```python
class Graph(object):
    def __init__(self,nodes):
        self.graph = {}
        self.distance = {}
        self.nodes = nodes
    
    def add_edge(self,u,v):
        self.graph.setdefault(u,[]).append(v)
    
    def bfs(self,s):
        distance = 0
        visited = [0]*(self.nodes+1)
        visited[s] = 1
        queue = []
        queue.append([s,distance])
        while queue:
            u = queue.pop()
            distance = u[1]+1
#             print(self.graph)
            for t in self.graph[u[0]]:
                if not visited[t]:
#                     if t == 0:
#                         print('here',t)
                    self.distance[t] = distance
                    queue.append([t,distance])
                    visited[t]=1
#         print(self.distance)
        for i in range(1,self.nodes+1):
            if not visited[i]:
#                 print(i)
#                 if not i:
#                     print('here',i)
                self.distance[i] = -1/6
#         sorted(mydict.items(), key=lambda item: item[1])
#         print(self.distance, 'num of nodes',self.nodes+1)
#         print((sorted(self.distance.items(), key = lambda x:x[0])))
        result = sorted(self.distance.items(), key = lambda x:x[0])
        result = [str(int(6*result[i][1])) for i in range(len(result))]
        
#         result = [6*i for i in self.distance.values()]
#         result.pop(s)
        return ' '.join(result)
    
        
def findShortest(graph_nodes, graph_from, graph_to, val):
    graph = Graph(graph_nodes)
    for u, v in zip(graph_from, graph_to):
        graph.add_edge(u, v)
        graph.add_edge(v, u)
    return graph.bfs(val)

for ha in range(1):
    graph_nodes, graph_edges = map(int, input().split())
    edges = list(map(int, input().split()))
    graph_from = edges[::2]
    graph_to = edges[1::2]
    val = int(input())
    ans = findShortest(graph_nodes, graph_from, graph_to, val)
    print(ans,'\nnext one\n')
```

```python
class Graph(object):
    def __init__(self,nodes):
        self.graph = {}
        self.distance = {}
        self.nodes = nodes
    
    def add_edge(self,u,v):
        self.graph.setdefault(u,[]).append(v)
    
    def bfs(self,s):
        distance = 0
        visited = [0]*(self.nodes+1)
        visited[s] = 1
        queue = []
        queue.append([s,distance])
        while queue:
            u = queue.pop()
            distance = u[1]+1
            for t in self.graph[u[0]]:
                if not visited[t]:
                    self.distance[t] = distance
                    queue.append([t,distance])
                    visited[t]=1
        for i in range(1,self.nodes+1):
            if not visited[i]:
                self.distance[i] = -1/6
        result = sorted(self.distance.items(), key = lambda x:x[0])
        result = [str(int(6*result[i][1])) for i in range(len(result))]
        return ' '.join(result)
    
        
def findShortest(graph_nodes, graph_from, graph_to, val):
    graph = Graph(graph_nodes)
    for u, v in zip(graph_from, graph_to):
        graph.add_edge(u, v)
        graph.add_edge(v, u)
    return graph.bfs(val)

t = int(input())
for i in range(t):
    graph_nodes, graph_edges = map(int, input().split())
    graph_from = []
    graph_to = []
    for i in range(graph_edges):
        x,y = [int(x) for x in input().split()]
        graph_from.append(x)
        graph_to.append(y)
    val = int(input())
    ans = findShortest(graph_nodes, graph_from, graph_to, val)
    print(ans)
```
