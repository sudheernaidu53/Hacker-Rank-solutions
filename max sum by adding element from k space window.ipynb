```python
def predator(arr):
    n = len(arr)
    roots = []
    prey = [[] for a in range(len(arr))]
    for i,val in enumerate(arr):
        if val!=-1:
            prey[val].append(i)
        else:
            roots.append(i)
    maxim = 0
    for root in roots:
        curr_tree = [root]
        queue = [root]
        while queue:
            curr_node = queue[-1]
            queue.pop()
            if prey[curr_node]:
                curr_tree.append(prey[curr_node])
                for k in prey[curr_node]:
                    queue.insert(0,k)
        maxim = max(maxim,len(curr_tree))
    return maxim
```
